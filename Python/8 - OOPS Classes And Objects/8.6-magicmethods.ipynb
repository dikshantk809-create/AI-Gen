## Magic Methods

Magic methods (also called dunder methods, because they have double underscores like __init__) are special methods in Python that let you define how objects of a class behave with built-in operations such as creating objects, printing them, comparing them, or using operators.

| Magic Method | Purpose                                                       |
| ------------ | ------------------------------------------------------------- |
| `__init__()` | Initializes an object (constructor).                          |
| `__str__()`  | Returns a human-readable string representation of the object. |
| `__repr__()` | Returns the official string representation of the object.     |
| `__len__()`  | Returns the length of an object using `len()`.                |
| `__add__()`  | Defines behavior for the `+` operator.                        |
| `__sub__()`  | Defines behavior for the `-` operator.                        |
| `__eq__()`   | Defines behavior for the `==` operator.                       |
| `__lt__()`   | Defines behavior for the `<` operator.                        |

```python
### Basics methods:
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
person=Person("John", 30)
print(person)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"name={self.name}, age={self.age}"
    
    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"
person=Person("John", 30)
print(person)
print(repr(person))
```
