#### Syntax and Semantics in Python

Syntax refers to the set of rules that defines the combinations of symbols that are considered to be correctly strctureed programs ina language. In simple terms, syntax is about the correct arrangement of words and symbols in a code.

Semantics refers to the meaning or the interpretation of the symbols, vharacters, and commands in a language. It is about what the  code is supposed to do when it run

```python
## Basic syntax Rules in  Python
# 1. Case Sensitivity: Python is case-sensitive, meaning that 'Variable' and 'variable' are considered different identifiers.
# 2. Indentation: Python uses indentation to define blocks of code. Consistent indentation is crucial for the code to run correctly.
name="dikshant"
Name="dishu"
print(name)
print(Name)
```

```python
# Indemtation:
## Python uses indentation to define blocks of code. Consistent indentation is crucial for the code to run correctly.

age=32
if age>30:
    print(age)
print(age)
```
