##### Deep Dive into Operators

```python
## Arithmethic Operation
a = 10
b = 3
add_result = a + b  # Addition
sub_result = a - b  # Subtraction
mul_result = a * b  # Multiplication
div_result = a / b  # Division
floor_div_result = a // b # Floor Division
mod_result = a % b  # Modulus
exp_result = a ** b  # Exponentiation
print("Addition:", add_result)
print("Subtraction:", sub_result)   
print("Multiplication:", mul_result)
print("Division:", div_result)
print("Floor Division:", floor_div_result)
print("Modulus:", mod_result)
print("Exponentiation:", exp_result)
# Output:
# Addition: 13
# Subtraction: 7    
# Multiplication: 30
# Division: 3.3333333333333335
# Floor Division: 3
# Modulus: 1
# Exponentiation: 1000



```

#### Logical Operators:

```python
## And , not, Or Operators
x=True
y=False
result_and = x and y
print(result_and)  # Output: False
```

```python
x=True
y=True
result_and = x and y
print(result_and)  # Output: True
```

```python
x=False
y=True
result_and = x and y
print(result_and)  # Output: False
```

```python
## Or Operator
x=False
y=True
result_or = x or y
print(result_or)  # Output: True
```

```python
x=True
y=True
result_or = x or y
print(result_or)  # Output: True
```

```python
x=False
y=False
result_or = x or y
print(result_or)  # Output: False
```

```python
## nOT Operator
x=True
result_not = not x
print(result_not)  # Output: False
```
