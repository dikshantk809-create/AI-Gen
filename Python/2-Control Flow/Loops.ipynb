```python
## For Loops:
# A for loop is used for iterating over a sequence (like a list, tuple, dictionary, set, or string).
for i in range(5):
    print(i)
```

```python
## String Iteration:
# You can also use a for loop to iterate through the characters of a string.
str="diskhant Aggarwal"
for i in str:
    print(i)
```

```python
## While Loops:
# A while loop repeatedly executes a target statement as long as a given condition is true.
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
count =0
while count%2==0:
    print(count)
    count +=1
```

```python
## loop control statements
# Loop control statements change execution from its normal sequence.
# They can be used to terminate a loop, skip to the next iteration, or jump to the loop's beginning.
# The loop control statements are:
# break
# continue
# pass
```

```python
## break statement
## The break statement is used to exit a loop prematurely when a certain condition is met.
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
## Continue Statement:
## The continue statement is used to skip the current iteration of a loop and move to the next iteration.
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
for i in range(10):
    if i%3==0:
        continue
    print(i)
```

```python
## Pass Statement:
## The pass statement is a null operation; it is used when a statement is syntactically
for i in range(5):
    if i ==3:
        pass
    print(i)
```

```python
## Nested Loops:
# A nested loop is a loop inside another loop. The "inner loop" will be executed
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
```

```python
## Examples- Calculate the sum of first N natural numbers using a while and for loop.

## while loop
n=10
sum=0
count=1
while count<=n:
    sum=sum+count
    count=count+1
print(f"The sum of first {n} natural numbers using while loop is: {sum}")
```

```python
n=10
sum=0
for i in range(11):
    sum=sum+i
print(f"The sum of first {n} natural numbers using for loop is: {sum}")
```

```python
## Examples-  prime numbers between 1 to 100 using for loop. 

for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
