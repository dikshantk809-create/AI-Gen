## File OPERATION-Read and Write Files

File handling is a crucial part of any programing lau=nguage.Python provides built-in functions and
methons to read from and write to files, both text and binary. This lesson will cover the basic of file handling, including
reading and writing text files and binary files.


```python
### Read a Whole file:
with open('example.txt','r') as file:
    content=file.read()
    print(content)
    
```

```python
## Read a file line by line
with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) 
```

```python
## Wriring a file(overwriting):
with open('example.txt','w') as file:
    file.write('hello world!\n')
    file.write("this is a new line")
```

```python
## write a file  without overwriting:
with open('example.txt','a') as file:
    file.write("Append operartion taking pl")
```

```python
### Writing a list of lines to a file:
lines=['first line\n','second line \n', 'third line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
### Binary Files:
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)

```

```python
## Reading a binary file:
with open('example.bin' , 'rb') as file:
    content = file.read()
    print(content)
```

```python
### Read the contect from source text file  amd write toa destintion text file:
with open('example.txt','r')as source_file:
    content = source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```

```python
# read a text file and count the number of lines, words and characters.
def count_txt_file(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        line_count =len(lines)
        word_count =sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'example.txt'
lines,words,characters = count_txt_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

```python
#Checking if a path is a file or directory:
import os 

path = 'example.txt'
if os.path.isfile(path):
    print(f"the path  '{path}' is a  file ")
elif os.path.isdir(path):
    print(f"the path  '{path}' is a directory.")
else:
    print(f"the path  '{path}' is neither a file nor a firectory.")
    
```
