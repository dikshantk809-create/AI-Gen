## Operator Overloading

Operator Overloading is a feature in object-oriented programming that allows programmers to redefine the behavior of operators (such as +, -, *, ==, etc.) for user-defined data types (classes or objects).

| **Operator** | **Magic Method**                | **Purpose**              | **Example**   |
| ------------ | ------------------------------- | ------------------------ | ------------- |
| `+`          | `__add__(self, other)`          | Addition                 | `a + b`       |
| `-`          | `__sub__(self, other)`          | Subtraction              | `a - b`       |
| `*`          | `__mul__(self, other)`          | Multiplication           | `a * b`       |
| `/`          | `__truediv__(self, other)`      | Division                 | `a / b`       |
| `//`         | `__floordiv__(self, other)`     | Floor Division           | `a // b`      |
| `%`          | `__mod__(self, other)`          | Modulus                  | `a % b`       |
| `**`         | `__pow__(self, other)`          | Exponent                 | `a ** b`      |
| `==`         | `__eq__(self, other)`           | Equal To                 | `a == b`      |
| `!=`         | `__ne__(self, other)`           | Not Equal To             | `a != b`      |
| `>`          | `__gt__(self, other)`           | Greater Than             | `a > b`       |
| `<`          | `__lt__(self, other)`           | Less Than                | `a < b`       |
| `>=`         | `__ge__(self, other)`           | Greater Than or Equal To | `a >= b`      |
| `<=`         | `__le__(self, other)`           | Less Than or Equal To    | `a <= b`      |
| `[]`         | `__getitem__(self, key)`        | Access Item              | `obj[0]`      |
| `[] =`       | `__setitem__(self, key, value)` | Set Item                 | `obj[0] = 10` |
| `len()`      | `__len__(self)`                 | Returns Length           | `len(obj)`    |
| `str()`      | `__str__(self)`                 | String Representation    | `str(obj)`    |
| `repr()`     | `__repr__(self)`                | Official Representation  | `repr(obj)`   |
| `()`         | `__call__(self)`                | Makes Object Callable    | `obj()`       |


```python
class Vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)

    def __sub__(self,other):
        return Vector(self.x-other.x,self.y-other.y)

    def __mul__(self,other):
        return self.x == other.x and self.y == other.y

    def __repr__(self):
        return f"Vector({self.x},{self.y})"

## create objects of the vector class :
v1=Vector(2,3)
v2=Vector(4,5)

print(v1+v2)
print(v1-v2)
print(v1*v2)

```
