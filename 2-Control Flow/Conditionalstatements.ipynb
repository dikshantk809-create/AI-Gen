#### Conditionalstatements (if,elif,else)

```python
## if statements:
## An if statement is used to test a specific condition. 
age=20
if age >=20:
    print("You are an adult.")
```

```python
## else statements:
## An else statement is used to execute a block of code when the condition in the if statement is false.
age=20
if age >=18:
    print("You are an adult.")  
else:
    print("You are a minor.")   
```

```python
age=int(input("enter your age:"))
if age >=18:
    print("You are an adult.")  
else:
    print("You are a minor.")   
```

```python
## elif statements:
## An elif statement is used to test multiple conditions in a sequence.
age=20
if age<13:
    print("You are a child.")
elif age<18:
    print("You are a teenager.")
else:
    print("You are an adult.")
```

```python
## Nested Conditional Statements:
## Nested conditional statements are if statements that are contained within another if statement.

## Example: number is even , odd , negative
num=int(input("enter a number:"))
if num>=0:
    print("The number is positive.")
    if num%2==0:
        print("The number is even.")
    else:
        print("The number is odd.")
else:
    print("The number is negative.")


```

```python
## Practice Problems:
## Determin if a year is a leap year  using nested if statements.
year=int(input("enter a year:"))
if year%4==0:
    if year%100==0:
        if year%400==0:
            print(f"{year}  is a leap year.")
        else:
            print(f"{year} is not a leap year.")
    else:
        print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")



```

```python
## make a simple calculator using if , elif , else statements.
num1=float(input("enter first number:"))
num2=float(input("enter second number:"))
operation=input("enter operation (+,-,*,/):")
if operation=='+':
    print(f"{num1} + {num2} = {num1+num2}")
elif operation=='-':
    print(f"{num1} - {num2} = {num1-num2}")
elif operation=='*':
    print(f"{num1} * {num2} = {num1*num2}")
elif operation=='/':
    if num2!=0:
        print(f"{num1} / {num2} = {num1/num2}")
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Invalid operation.")
```
