### Dictionaries

#### Introduction to Dictionaries:

Dictionaries are unordered colllections of items. They store data in key-value pairs. keys must be unique and 
immmutable(eg., strings,numbers, tuples),while values can be of any type.

```python
#### Creating a Dictionary
# This script creates a dictionary with sample data and prints it.
empty_dict = {}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"dikshant","age":21,"grade":24}
print(student)
print(type(student))
```

```python
### Single key is always used
student={"name":"dikshant","age":21,"name":24}
print (student)
```

```python
## Accessing Dictionary Elements:
student={"name":"dikshant","age":21,"grade":"A"}
student
```

```python
### Accessing individual elements
print(student["name"])
print(student["age"])
print(student["grade"])
```

```python
## Accessing using get() method:
print(student.get("name"))
print(student.get("last_name"))
print(student.get("last_name","not found"))
```

```python
## Modification of Dictionary Elements:
## Dictionary is mutable so we can change the values of existing keys.
student={"name":"dikshant","age":21,"grade":"A"}
student
```

```python
student["age"]=33 ## Modifying age from 21 to 33
print(student)
student["address"]="India"  ## Adding new key-value pair
print(student)
```

```python
del student["grade"]    ## Deleting key-value pair with key 'grade'
print(student)
```

```python
## Dictionary Methods:
student={"name":"dikshant","age":21,"grade":"A"}
keys=student.keys()     ### Getting all the keys
print(keys)
values=student.values()  ### Getting all the values
print(values)

items=student.items()  ### Getting all key-value pairs
print(items)
```

```python
## Shallow copy of Dictionary:
student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="dikshant Aggarwal"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy()  ## Using copy() method to create a shallow copy
print(student)
print(student_copy1)
```

```python
student["name"]="dikshant Aggarwal2"
print(student_copy1)
print(student)
```

```python
#### Iterating Over a Dictionary:
### you can iterate over the keys, values, or key-value pairs in a dictionary using loops.

### Iterating over keys:
for keys in student.keys():
    print(keys)

```

```python
### Iterating over values:
for values in student.values():
    print(values)
```

```python
## Iterating over key-value pairs:
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
### Nested Dictionaries:
## A nested dictionary is a dictionary that contains other dictionaries as its values.
nested_dict={
    "student1":{"name":"dikshant","age":21},
    "student2":{"name":"rahul","age":22},
    "student3":{"name":"anita","age":20}
}
print(nested_dict)
```

```python
## Accessing elements in nested dictionaries:
print(nested_dict["student1"]["name"])
print(nested_dict["student1"]["age"])
```

```python
nested_dict.items()
```

```python
## Iterating over nested dictionaries:
for student_id,student_info in nested_dict.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"  {key}:{value}")
```

```python
## Dictionary Comprehensions:
squares={x:x**2 for x in range(0,6)}
print(squares)
```

```python
## Condition dictionary comprehension:
evens={x:x**2 for x in range(0,11) if x%2==0}
print(evens)
```

```python
### Practical Exercise:
# Create a dictionary that maps numbers from 1 to 10 to their cubes using dictionary comprehension

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)

```

```python
## Merge Dictionaries:
dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
