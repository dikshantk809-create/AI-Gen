## Understanding Exceptions:-

Exception handlind=g in Python allows you to handle errors gracefully and take 
corrective action actions without stopping the excution of the program. this lesson will
cover the basics of exceptions, including how to use try, except else, and finally blocks.

What are exceptions?

Exceptions are events that disrupt the normal flow of a program,they occur when an error is encountered
during program execution.Common exceptions include:


```python
## Exception  try except block:
try:
    a=b
except:
    print("the variable b is not defined")
```

```python
try:
    a=b
except NameError as ex:
    print(ex)
```

```python
try:
    result=1/0
except ZeroDivisionError as ex:
    print(ex)
    print("You cannot divide by zero")
```

```python
try:
    result=1/2
    a=b
except ZeroDivisionError as ex:
    print(ex)
    print("You cannot divide by zero")
except Exception as ex1:
    print(ex1)
```

```python
try:
    num=int(input("enter the number"))
    result=10/num
except ValueError as ex:
    print("this is not a number")
except ZeroDivisionError as ex:
    print("You cannot divide by zero")
except Exception as ex1:
    print(ex1)
```

```python
## try,except,else block:
try:
    num=int(input("enter the number"))
    result=10/num
except ValueError as ex:
    print("this is not a number")
except ZeroDivisionError as ex:
    print("You cannot divide by zero")
except Exception as ex1:
    print(ex1)
else:
    print("The result is:", result)
```

```python
## try,except,else block and finally block:
try:
    num=int(input("enter the number"))
    result=10/num
except ValueError as ex:
    print("this is not a number")
except ZeroDivisionError as ex:
    print("You cannot divide by zero")
except Exception as ex1:
    print(ex1)
else:
    print("The result is:", result)
finally:
    print("This block will always execute")
```
