#### Data Types

1. Definition:
    >Data types arw a classification of data which tell the compiler or interpreter how the programmer intends to use the data.

    >They determine the type of operations that can be performed on the data, the values that the data can take, and tha amount of memory needed to store the data.

2. Importance od data types in programming:

>Data Types ensure that data is stored in an efficient way.
>They help in performing correct operations on data.
>proper use of data types can prevent errors and bugs in the program

```python
## Integer Example:
age = 25
type(age)  # <class 'int'>
```

```python
## Floating Point Example:
height = 5.9
print(height)  # 5.9
print(type(height))  # <class 'float'>
```

```python
## string Example:
name = "Dikshant"
print(name)  # Dikshant
print(type(name))  # <class 'str'>
```

```python
## boolean Example:
is_student = True
print(is_student)  # True   
print(type(is_student))  # <class 'bool'>
```

```python
a=10
b=10
type(a==b)
```

```python
## Common error:
result="hello" + 5  # This will raise a TypeError
```
