#### Tuples

Tuples are ordered collections of items that are immutable. They are similar to lists , but their immutablility makes them different.

```python
-## Creating a tuple:
empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,4,5])
numbers
```

```python
list((1,2,3,4,5))
```

```python
mixed_tuple=(1,"hello World",3.14,True)
print(mixed_tuple)
```

```python
## Accessing Tuple Elements:
numbers
```

```python
numbers[2]
numbers[-1]
```

```python
numbers[0:4]
```

```python
numbers[::]
```

```python
numbers[::-2]
```

```python
## Tuple Operations:
# Concatenation
concatenated_tuple=numbers + mixed_tuple
print(concatenated_tuple)
```

```python
mixed_tuple * 3
```

```python
numbers *3
```

```python
## Immutable Nature of Tuples:
## Trying to change an element (This will raise an error)
# numbers[0]=10  # Uncommenting this line will raise a TypeError
lst=[1,2,3,4,5]
print(lst)

lst[1]="dikshant"
print(lst)
```

```python
numbers[1]="dishu"
```

```python
numbers
```

```python
## Tuple Methods:
print(numbers.count(1))
print(numbers.index(4))
```

```python
## Packing and Unpacking Tuples: 
packed_tuple=1,"hello",3.14
print(packed_tuple)
```

```python
## Unpacking a tuple:
a,b,c=packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuples:
## Example of a nested list
lst=[[1,2,3,4],[5,6,7,8,9],[1,"hello",3.14,"c"]]
lst[0][2]
lst[0][0:3]
```

```python
lst=[[1,2,3,4],[5,6,7,8,9],(1,"hello",3.14,"c")]
lst[2][2]
lst[2][0:3]
```

```python
nested_tuple =((1,2,3),("a","b","c"),(True,False))
## Accessing elements in nested tuples
print(nested_tuple[0]) 
print(nested_tuple[1][2]) 
```

```python
## Iterating over nested tuples:
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
