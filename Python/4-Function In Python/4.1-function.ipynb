## Introduction to Functions:

Definition:-


A Function is a block of code that prefroms a specific task. Functions help in organizing code,reusing code  and improving readability.

```python
## Syntax:
def function_name(parameters):
    """Docstring"""
    # Function Body
    return expression
```

```python
## why functions?
num=23
if num%2==0:
    print(f"{num} is even ")
else:
    print(f"{num} is odd ")
```

```python
def even_or_odd(num):
    """This fnction finds even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
```

```python
## Call this functions:
even_or_odd(34)
```

```python
### Function with multiple parameters:
def add(a,b):
    """"This function adds two numbers"""
    return a+b
result=add(4,5)
print(result)  
```

```python
## Default parameters:
def greet(name):
    print(f"hello {name} Welcome to the paradise")
greet("dikshant ")
```

```python
def greet(name):
    print(f"hello {name} Welcome to the paradise")
greet()
```

```python
def greet(name="guest"):
    print(f"hello {name} Welcome to the paradise")
greet()
```

```python
### Varible Length Arguments:
## Positional Arguments (*args);
def print_numbers(*args):
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"dikshant")
```

```python
### Keyword Arguments ( **kwargs);
def print_details(**kwargs):
    for key,values in kwargs.items():
        print(f"{key} : {values}")
```

```python
print_details(name="dikshant",age=21,city="derabassi")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f"Positional argument: {val}")
    for key,values in kwargs.items():
        print(f"{key} : {values}")
```

```python
print_details(1,2,3,name="dikshant",age=21,city="derabassi")
```
