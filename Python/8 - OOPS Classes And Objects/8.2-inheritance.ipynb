## Inheritance In Python


nheritance is an Object-Oriented Programming (OOP) feature in Python that allows one class (called the child class or derived class) to inherit the properties (attributes) and methods of another class (called the parent class or base class).

```python
## Inheritance
## Parent class

class Car:
    def __init__(self,windows,doors,enginetype):
        self.windows=windows
        self.doors=doors
        self.enginetype=enginetype
    
    def drive(self):
        print(f"the person will drive the {self.enginetype} car")
```

```python
car1=Car(4,5,"petrol")
print(car1.drive())
```

```python
class Tesla(Car):
    def __init__(self,windows,doors,enginetype,is_selfdriving):
        super().__init__(windows,doors,enginetype)
        self.is_selfdriving=is_selfdriving

    def selfdriving(self):
        print(f"Tesla supports {self.is_selfdriving} self driving")



```

```python
tesla1=Tesla(4,5,"electric",True)
print(tesla1.selfdriving())
```

```python
tesla1.drive()
```

```python
### Multiple Inheritance
## when a class inherits from more than one class, it is called multiple inheritance.
#  The child class inherits the properties and methods of all the parent classes.
## basa class 1

class Animal:
    def __init__(self,name):
        self.name=name

    def speak(self):
        print("Animal speaks")

## base class 2 
class Pet:
    def __init__(self,owner):
        self.owner=owner

## Derived class
class Dog(Animal,Pet):
    def __init__(self,name,owner):
        Animal.__init__(self,name)
        Pet.__init__(self,owner)

    def speak(self):
        return f"{self.name} say woof"
## create an object
dog=Dog("Buddy","John")
print(dog.speak())
print(f"Qwner of the dog is {dog.owner}")
```
