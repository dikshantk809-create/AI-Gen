## Polymorphism

Polymorphism is an object-oriented programming (OOP) concept that allows one interface or method to have multiple forms or behaviors. In other words, the same method or function can perform different tasks depending on the object that calls it.

```python
## Base class:
class Animal:
    def speak(self):
        return "Animal speaks"

## Derived class:
class Dog(Animal):
    def speak(self):
        return "Dog barks"


## Derived class
class Cat(Animal):
    def speak(self):
        return "Cat meows"

## Function that demonstrates polymorphism:
def animal_speak(animal):
    print(animal.speak())


dog=Dog()
print(dog.speak())

cat=Cat()
print(cat.speak())


animal_speak(dog)
```

```python
## Polymorphissm with functions and methods:
## base class
class Shape:
    def area(self):
        return "the area is calculated"

## Derived class
class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height

    def area(self):
        return self.width*self.height


## Derived class 2 
class Circle(Shape):
    def __init__(self,radious):
        self.radius=radious

    def area(self):
        return 3.14*self.radius*self.radius

## Function that demonstrates polymorphism:
def print_area(shape):
    print(shape.area())


rectangle=Rectangle(4,5)
circle=Circle(3)

print_area(rectangle)
print_area(circle)
```

```python
## Polymorphism with Abstract Base Classes (ABC):-
from abc import ABC,abstractmethod
## Define an abstract class
class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

## Derived class
class Car(Vehicle):
    def start(self):
        return "Car started"
    
## Derived class2:
class Motorcycle(Vehicle):
    def start(self):
        return "Motorcycle started"
    
### Function that demonstrates polymorphism:
def start(vehicle):
    print(vehicle.start())  

#create objects of derived classes:
car=Car()
motorcycle=Motorcycle()
start(car)
```
